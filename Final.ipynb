```python
c3 = final.solve_odes(3)      #c = 3
c4 = final.solve_odes(4)      #c = 4
c415 = final.solve_odes(4.15) #c = 4.15
c42 = final.solve_odes(4.2)   #c = 4.2
c57 = final.solve_odes(5.7)   #c = 5.7
```

```python
final.plotx(c3)
final.plotx(c4)
final.plotx(c415)
final.plotx(c42)
final.plotx(c57)
```

## Analysis
When C = 2, the oscilations are very consistant. The graph grows steadily until it reaches a consistant oscilatiion around t = 50. However, as the C value increases, the more sparatic and the graphs become. Not only do the local max and mins increase in size, but the number of wavelengths in one oscilation increases as well until it is hard to even tell where one starts and one ends. 

```python
final.plotxy(c3)
final.plotxy(c4)
final.plotxy(c415)
final.plotxy(c42)
final.plotxy(c57)
```

## Explanation 
The 2D graph starts as a circle, or more two loops that start and end at the same place. As C increases, the number of loops increases along with the size of the loops. The higher C goes, the more unpredictable those loops become. However, the fact that they all start and end in the same place references the oscilations in the x(t) graph. 

```python
final.plotxyz(c3)
final.plotxyz(c4)
final.plotxyz(c415)
final.plotxyz(c42)
final.plotxyz(c57)
```

## Explanation
The 3D graph is very similar to the 2D one, except in 3D. What starts as rings that link grows as the C value grows. As it grows it gain larger, more speratic loops. The difference is you can see the upward values of the circle which were invisible in the 2D plot. The more rings there are, the higher they go, and wider they span.
